```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
data = pd.read_csv('Historical Options Chain.csv')
data.head()
```

```python
previous = data[-40:-20]
most_recent = data[-20:]
prev_strikes = []
for x in previous['Strike']:
    prev_strikes.append(x)
print(prev_strikes)
for _ in prev_strikes:
    test = most_recent[most_recent['Strike']==_]
    print(test)
```
